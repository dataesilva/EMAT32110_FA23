```python
import sqlite3
import pandas as pd
```

[Database Diagram](https://www.sqlitetutorial.net/wp-content/uploads/2018/03/sqlite-sample-database-diagram-color.pdf)

[Download the example database](https://www.sqlitetutorial.net/wp-content/uploads/2018/03/chinook.zip) and unzip.

```python
# check path, not on GitHub repo
chinook = sqlite3.connect(r'D:\chinook.db')
```

```python
dims_albums = '''
SELECT *
FROM albums
LIMIT 1
'''
```

```python
dims_df = pd.read_sql_query(dims_albums, chinook)
dims_df
```

```python
query_albums = '''
SELECT AlbumId, Title, ArtistId
FROM albums
'''
```

```python
albums_df = pd.read_sql_query(query_albums, chinook)
albums_df.shape
```

```python
albums_df.head()
```

```python
query_a_albums = '''
SELECT *
FROM albums
WHERE Title LIKE 'A%'
'''
```

```python
a_albums_df = pd.read_sql_query(query_a_albums, chinook)
a_albums_df.shape
```
