```python
import numpy as np
import pandas as pd
from statsmodels.formula import api as sm
from stargazer.stargazer import Stargazer
from IPython.core.display import HTML
```

```python
ces = pd.read_csv('in-class_notebooks/cces_example_data.csv')
ces.head()
```

```python
ces.columns
```

```python
model_1 = sm.ols("Symb_Threat ~ age + faminc_new + ideo5", data = ces)
```

```python
model_1.fit().summary()
```

```python
results_1 = model_1.fit()
model_1_table = Stargazer([results_1])
HTML(model_1_table.render_html())
```

```python
ces['pid3_labeled'] = ces['pid3'].replace([1,2,3], ['Dem', 'Rep', 'Ind'])
```

```python
model_2 = sm.ols("LibMediaUse ~ C(pid3_labeled)", data = ces)
```

```python
model_2.fit().summary()
```
