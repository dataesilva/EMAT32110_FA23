```python
query_album_artists = '''
SELECT Title, Name
FROM albums a
INNER JOIN artists r
ON (r.ArtistId = a.ArtistId)
'''
```

```python
data_album_artists = pd.read_sql_query(query_album_artists, chinook)
```

```python
q_track_album = '''
SELECT Name, Title, Milliseconds
FROM tracks t
INNER JOIN albums a
ON (t.AlbumId = a.AlbumId)
LIMIT 100
'''
```

```python
data_track_album = pd.read_sql_query(q_track_album, chinook)
```

```python
data_track_album
```

```python
data_track_album.groupby('Title').count()
```

```python
q_track_album_artist = '''
SELECT t.Name as Track,
        Milliseconds as TrackMS,
        Title as Album,
        r.Name as Artist
FROM tracks t
INNER JOIN albums a
ON (t.AlbumId = a.AlbumId)
INNER JOIN artists r
ON (r.ArtistId = a.ArtistId)
WHERE Track LIKE "D%"
LIMIT 100
'''
```

```python
data_taa = pd.read_sql_query(q_track_album_artist, chinook)
```
