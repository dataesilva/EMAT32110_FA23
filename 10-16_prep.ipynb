```python
ones = ces[ces['pid3'] == 1]['Symb_Threat']
twos = ces[ces['pid3'] == 2]['Symb_Threat']
threes = ces[ces['pid3'] == 3]['Symb_Threat']

fig2, ax2 = plt.subplots()
ax2.boxplot(ones, positions = [1], widths = [.75])
ax2.boxplot(twos, positions = [2], widths = [.75])
ax2.boxplot(threes, positions = [3], widths = [.75])
plt.show()
```

```python
ces.boxplot('Symb_Threat', by = 'pid3')
plt.show()
```
