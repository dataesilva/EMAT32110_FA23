```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
audio_features = pd.read_csv('audio_features.csv')
audio_features.head()
```

```python
audio_features.columns
```

```python
fig1, ax1 = plt.subplots(1, 2, figsize = (7, 5))
ax1[0].hist(audio_features['valence'])
ax1[1].hist(audio_features['danceability'])

plt.show()
```

```python
fig2, ax2 = plt.subplots()
ax2.scatter(x = audio_features['danceability'],
            y = audio_features['valence'])

plt.show()
```

```python
fig3, ax3 = plt.subplots()
ax3.scatter(x = audio_features['energy'],
            y = audio_features['instrumentalness'], alpha = .5)

plt.show()
```

```python
fig4, ax4 = plt.subplots()
ax4.scatter(x = audio_features['energy'],
            y = audio_features['loudness'], alpha = .5)

plt.show()
```

```python
audio_features[audio_features['album'] != 'midnights']
```

```python
mid_df = audio_features[audio_features['album'] == 'midnights']
lon_df = audio_features[audio_features['album'] == 'loneliest']
fig5, ax5 = plt.subplots()
ax5.scatter(x = mid_df['danceability'],
            y = mid_df['valence'],
            label = 'Midnights',
           color = 'red',
           marker = 'd',
           alpha = .5)
ax5.scatter(x = lon_df['danceability'],
            y = lon_df['valence'],
            label = 'The Loneliest Time',
           color = 'green',
           marker = 's',
           alpha = .5)

ax5.set_title('Relationship between\nDanceability and Valence for Two 2022 Pop Albums')
ax5.set_xlabel('Danceability')
ax5.set_ylabel('Valence')
ax5.legend()
plt.show()
```
