```python
Client_Secret = pd.read_csv(r"E:\Dropbox\Kent\EMAT-32110_Data-in-EMAT\Spotify-Keys_9-12-23.txt")['Client_Secret'].iloc[0]
```

```python
# Client_ID
```

```python
client_cred = base64.b64encode(str(Client_ID + ":" + Client_Secret).encode("ascii"))
```

```python
# client_cred
```

```python
headers = {"Authorization": "Basic {}".format(client_cred.decode("ascii"))}
```

```python
# headers
```

```python
payload = {'grant_type' : 'client_credentials'}
url = 'https://accounts.spotify.com/api/token'
```

```python
session_key_response = requests.post(url = url, data = payload, headers = headers)
```

```python
session_key_response.status_code
```

```python
# session_key_response.json()['access_token']
```

```python
session_header_key = session_key_response.json()
```

```python
key = session_header_key['access_token']
```

```python
session_headers = {"Authorization": "Bearer {}".format(key)}
```

```python
# session_headers
```

```python
urllib.parse.quote('this is')
```

```python
search_url = 'https://api.spotify.com/v1/search?q={}&type={}&limit=50'.format(urllib.parse.quote('autumn'),'track')
```

```python
search_result = requests.get(search_url, headers = session_headers)
```

```python
autumn_tracks_results = search_result.json()
```

```python
autumn_tracks_results.keys()
```

```python
len(autumn_tracks_results['tracks']['items'])
```

```python
autumn_tracks_results['tracks']['items'][0].keys()
```

```python
autumn_tracks_df = pd.DataFrame(autumn_tracks_results['tracks']['items'])
```

```python
autumn_tracks_df.head()
```

```python
first_artist = [x['artists'][0]['name'] for x in autumn_tracks_results['tracks']['items']]
```

```python
autumn_tracks_df['first_artist'] = first_artist
```

```python
urllib.parse.quote(','.join(list(autumn_tracks_df['id'])))
```

```python
track_audio_features_url = 'https://api.spotify.com/v1/audio-features?ids={}'.format(urllib.parse.quote(','.join(list(autumn_tracks_df['id']))))
```

```python
features_response = requests.get(url = track_audio_features_url, headers = session_headers)
```

```python
track_features = features_response.json()
```

```python
track_features_df = pd.DataFrame(track_features['audio_features'])
```

```python
track_features_df.head()
```

```python
full_data_df = pd.merge(autumn_tracks_df, track_features_df, how = 'inner', on = 'id')
```

```python
full_data_df.head()
```

```python
full_data_df.columns
```
